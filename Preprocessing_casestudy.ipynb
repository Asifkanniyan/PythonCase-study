```python
## 5. Check and handle outliers in at least 3 columns in the dataset (Age)
```

```python
plt.boxplot(data['Age'])
plt.show()
```

```python
Q1 = np.percentile(data['Age'],25)
Q1
```

```python
Q3 = np.percentile(data['Age'],75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
lower_limit
```

```python
Upper_limit = Q3 + 1.5*IQR
Upper_limit
```

```python
outlier = []
for i in data['Age']:
  if (i < lower_limit) or (i > Upper_limit):
    outlier.append(i)
print(outlier)
```

```python
ind = data['Age']>Upper_limit
data.loc[ind].index
```

```python
data.drop(data.loc[ind].index, inplace = True)
```
